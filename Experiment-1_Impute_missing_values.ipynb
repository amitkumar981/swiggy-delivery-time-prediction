```python
cols_to_drop=['latitude', 'longitude','delivery_location_latitude', 'delivery_location_longitude',
'day', 'month','order_time_hr']
df.drop(columns=cols_to_drop,inplace=True)
df.shape
```

```python
df.columns
```

```python
df.isnull().any(axis=0).loc[lambda x: x].index
```

```python
num_cols=['person_age','person_ratings','pickup_time_in_minutes','distance']
num_cat_cols=['weather_condition','multiple_deliveries','festival','city','time_slot','vehicle_condition', 'order_type', 'vehicle_type','city_name',
'day_of_week']
ordinal_cat_cols=['traffic_density','distance_type']
```

```python
df['distance_type'].unique()
```

```python
traffic_order=['Low ','Medium ','Jam ','High ']
distance_order=['short','medium','long','very long']
```

```python
#saprate dependent and independent variable
x=df.drop(columns='time_taken')
y=df['time_taken']
```

```python
#spliting
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
x_train.shape,x_test.shape
```

```python
#features to fill mode
feature_to_fill_mode=['multiple_deliveries','festival','city']
features_to_fill_missing=[col for col in num_cat_cols if col not in feature_to_fill_mode]
features_to_fill_missing
```

```python
simple_imputer=ColumnTransformer(transformers=[
    ('mode_imputer',SimpleImputer(strategy='most_frequent'),feature_to_fill_mode),
    ('missing_imputer',SimpleImputer(strategy='constant',fill_value='missing'),features_to_fill_missing)
],remainder="passthrough",n_jobs=-1,force_int_remainder_cols=False,verbose_feature_names_out=False)
simple_imputer
simple_imputer.set_output(transform='pandas')
```

```python
simple_imputer.fit_transform(x_train).isna().sum()
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('scaling',MinMaxScaler(),num_cols),
    ('OHE',OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False),num_cat_cols),
    ('Ordinal encoding',OrdinalEncoder(categories=[traffic_order,distance_order],encoded_missing_value=-999,
                                       handle_unknown='use_encoded_value',unknown_value=-1),ordinal_cat_cols)
],remainder="passthrough",n_jobs=-1,force_int_remainder_cols=False,verbose_feature_names_out=False)
preprocessor.set_output(transform='pandas')
```

```python
preprocessor.fit_transform(x_train).isna().sum().loc[lambda ser :ser.ge(1)]
```

```python
knn_imputer=KNNImputer(n_neighbors=5)
```

```python
preprocessing_pipline=Pipeline(steps=[
    ('simple_imputer',simple_imputer),
    ('preprocessing',preprocessor),
    ('knn_imputer',knn_imputer)
])
preprocessing_pipline
```

```python
rf=RandomForestRegressor(verbose=True)
```

```python
model_pipe=Pipeline(steps=[
    ('preprocessing_pipeline',preprocessing_pipline),
    ('randomForest Regressor',rf)
])
model_pipe
```

```python
pt=PowerTransformer()
y_train_pt=pt.fit_transform(y_train.values.reshape(-1,1))
y_test_pt=pt.transform(y_test.values.reshape(-1,1))
```

```python
model_pipe.fit(x_train,y_train_pt.ravel())
```

```python
y_pred_train=model_pipe.predict(x_train)
y_pred_test=model_pipe.predict(x_test)
```

```python
y_pred_train_org=pt.inverse_transform(y_pred_train.reshape(-1,1))
y_pred_test_org=pt.inverse_transform(y_pred_test.reshape(-1,1))
```

```python
print('training_error',mean_absolute_error(y_train,y_pred_train_org))
print('test_error',mean_absolute_error(y_pred_test_org,y_test))
```

```python
print('training r2_score',r2_score(y_train,y_pred_train_org))
print('test_r2_score',r2_score(y_test,y_pred_test_org))
```
